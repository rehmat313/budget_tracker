```python
# ============================================
# WIKIPEDIA AGENT WITH GEMINI - IMPROVED VERSION
# ============================================

# Step 1: Install packages (run once)
!pip -q install google-genai wikipedia

# Step 2: Imports
from google import genai
from google.colab import userdata
import wikipedia

# Step 3: Get API key from Colab Secrets
# Make sure you saved GEMINI_API_KEY in ðŸ”‘ Secrets panel
client = genai.Client(api_key=userdata.get('GEMINI_API_KEY'))

# Step 4: Wikipedia Tool
def search_wikipedia(query):
    try:
        return wikipedia.summary(query, sentences=3)
    except:
        return "No result found."

# Step 5: Decision Brain (SEARCH or ANSWER only)
def think(question):
    response = client.models.generate_content(
        model="gemini-3-flash-preview",
        contents=f"""
You are a decision-making AI agent.

If the question requires looking up information, respond with ONLY:

SEARCH

If you already know the answer confidently, respond with ONLY:

ANSWER

Do not explain. Only respond with one word.

Question: {question}
"""
    )

    return response.text.strip()

# Step 6: Main Agent Function
def agent(question):
    print("="*50)
    print("GOAL:", question)
    print("="*50)

    decision = think(question)
    print("\nDecision:", decision)

    # If search is needed
    if decision == "SEARCH":
        print("\nUsing Wikipedia tool...\n")
        info = search_wikipedia(question)

        final_response = client.models.generate_content(
            model="gemini-3-flash-preview",
            contents=f"""
Use the information below to answer clearly.

Question: {question}

Information:
{info}
"""
        )

        return final_response.text

    # If model already knows answer
    else:
        final_response = client.models.generate_content(
            model="gemini-3-flash-preview",
            contents=f"Answer clearly and concisely:\n\n{question}"
        )

        return final_response.text

# Step 7: Test the Agent
print(agent("Is there any recent conflict b/w Pakistan and Afghanistan"))
```
